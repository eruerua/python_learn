```python
import tushare as ts
a_basic=ts.get_stock_basics()
```

```python
gpr=a_basic['gpr'].tolist()
npr=a_basic['npr'].tolist()
gpr_mean=round(sum(gpr)/float(len(gpr))/100,2)
npr_mean=round(sum(npr)/float(len(npr))/100,2)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Liquid, Gauge,WordCloud,Line


gpr_lq = (
        Liquid()
        .add("", [gpr_mean],is_outline_show=False)
        .set_global_opts(title_opts=opts.TitleOpts(title="gpr平均值"))
)
gpr_lq.render_notebook()
```

```python
npr_lq = (
        Liquid()
        .add("", [npr_mean],is_outline_show=False)
        .set_global_opts(title_opts=opts.TitleOpts(title="npr平均值"))
)
npr_lq.render_notebook()
```

```python
gpr_ga = (
        Gauge()
        .add("", [("gpr平均值",gpr_mean*100)])
        .set_global_opts(title_opts=opts.TitleOpts(title="gpr平均值"))
)
gpr_ga.render_notebook()
```

```python
npr_ga = (
        Gauge()
        .add("", [("npr平均值",npr_mean*100)])
        .set_global_opts(title_opts=opts.TitleOpts(title="npr平均值"))
)
npr_ga.render_notebook()
```

```python
a_industry=a_basic.groupby('industry')['name'].agg({'各行业公司数量':'count'})
a_industry.head()
```

```python
industry_count=[list(i) for i in zip(a_industry.index.tolist(),a_industry['各行业公司数量'].tolist())]
industry_wc = (
        WordCloud()
        .add("", industry_count,shape='circle')
        .set_global_opts(title_opts=opts.TitleOpts(title="各行业公司数量"))
)
industry_wc .render_notebook()
```

```python
maotai= ts.get_hist_data('600519')
maotai.head()
```

```python
maotai_date=maotai.index.tolist()[::-1]
maotai_high=maotai['high'].tolist()[::-1]
maotai_low=maotai['low'].tolist()[::-1]
line= (
        Line(init_opts=opts.InitOpts(width="900px", height="500px"))
            .add_xaxis(maotai_date)
            .add_yaxis("最高价", maotai_high, label_opts=opts.LabelOpts(position="top"))
            .add_yaxis("最低价", maotai_low, label_opts=opts.LabelOpts(position="bottom"))
            .set_global_opts(title_opts=opts.TitleOpts(title="茅台最高价最低价曲线"),
                             yaxis_opts=opts.AxisOpts(min_=400))
    )
line.render("茅台最高价最低价曲线.html")
```
