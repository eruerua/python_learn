```python
import pandas as pd
f = open('task_2_lianjia_data.csv',encoding='UTF-8') 
lianjia_data = pd.read_csv(f)
f.close()
```

```python
lianjia_data.head(3)
```

```python
#分组聚合
lianjia_data_gb_area=lianjia_data.groupby('面积大小（㎡）').agg('count')
```

```python
#转换为列表
area_list=lianjia_data_gb_area.index.tolist()
area_count_list=lianjia_data_gb_area['区'].tolist()
```

```python
##画图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False#显示中文
plt.title('area and count')
plt.plot(area_list, area_count_list, 'bo-')
plt.show()
plt.title('area and count')
plt.bar(area_list, area_count_list, alpha=0.9, width =1, facecolor = 'red', edgecolor = 'red', label='one', lw=1)
plt.show()
```

由图可见，面积大小在25-50区间的房源较多，其余房源较少

```python
##按照区显示房源数量
lianjia_data_gb_location=lianjia_data.groupby('区').agg('count')
location_list=lianjia_data_gb_location.index.tolist()
location_count_list=lianjia_data_gb_location['村'].tolist()
plt.title('location and count')
plt.bar(location_list, location_count_list, alpha=0.9, width =1, facecolor = 'red', edgecolor = 'red', label='one', lw=1)
plt.show()
```

由条形图可以看出，奉贤、嘉定等地房源数量较少，徐汇、静安房源数量较多。

```python
#按照楼层类型显示房源数量
lianjia_data_gb_floor=lianjia_data.groupby('楼层类型').agg('count')
floor_list=lianjia_data_gb_floor.index.tolist()
floor_count_list=lianjia_data_gb_floor['村'].tolist()
plt.title('floor and count')
plt.pie(floor_count_list, labels=floor_list,)
plt.show()
```

由图可见，高楼层房源数量较多，地下室房源较少。
